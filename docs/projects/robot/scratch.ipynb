```python
%%host

fs = 100
tau = 0.1
a = 1/(1+tau*fs)

def lpf(x, y_last):
    return y_last + a*(x-y_last)
```

```python
%%host

y = 0
for x in range(20):
    y = lpf(2, y)
    print(y)
```
